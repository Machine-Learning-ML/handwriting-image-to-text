```python
import cv2
import pytesseract
from matplotlib import pyplot as plt
```

```python
# Paths are relative to the project root (this folder)
image_path = "image/handnote-10.png"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
print("Grayscale image:")
# cv2_imshow(gray)
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
plt.title("Original image")
plt.axis("off")
plt.show()
```

```python
extracted_text = pytesseract.image_to_string(image_rgb)
print("Extracted text:\n")
print(extracted_text)
```

```python
import cv2
import pytesseract

image = cv2.imread("image/handnote-11.png")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Set Tesseract path if needed (especially on Windows)
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

text = pytesseract.image_to_string(thresh)
print("Extracted text:")
print(text)
```
